üì¶ Importing libraries that I will need for webscraping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

üõ†Ô∏è Set URL of website in a variable

```python
url = 'https://www.scrapethissite.com/pages/forms/?page_num=16'
page = requests.get(url)
page
```

üç≤ Use BeautifulSoup to extract the HTML content

```python
soup = BeautifulSoup(page.text, 'html')
print(soup)
```

üìã Extract the table with the Hockey Scores

```python
hockey_table = soup.find_all('table', class_='table')
print(hockey_table)
```

üìù Extract the column headings

```python
table_titles = soup.find_all('th')
hockey_table_title = [title.text.strip() for title in table_titles]
print(hockey_table_title)
```

üîç Find the first table (adjust if needed)

```python
table = soup.find('table')
```

üìÑ Get all rows in the table

```python
rows = table.find_all('tr')
```

üîÅ Loop through and print the first 25 data rows (excluding header)

```python
for i, row in enumerate(rows[0:26], start=1):  # skip header (index 0)
    cells = row.find_all(['td', 'th'])
    row_data = [cell.get_text(strip=True) for cell in cells]
    print(f"Row {i}:", row_data)
```

üíæSave the column headings onto a Pandas DataFrame

```python
df = pd.DataFrame(columns=hockey_table_title)
df
```
